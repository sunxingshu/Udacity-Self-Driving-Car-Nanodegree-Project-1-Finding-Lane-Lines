## Helper Functions

Here are some function provided to faciliate the project. Some of the codes are modified to enhance the algorithm.

```python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    `vertices` should be a numpy array of integer points.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, s_flag_l, s_flag_r ,color=[255, 0, 0], thickness=20):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    imshape = img.shape
    left_line = np.empty((2,0))
    right_line = np.empty((2,0))
    if lines is not None:
        for line in lines:
            for x1,y1,x2,y2 in line:
                if ((y1-y2)/(x1-x2+1e-4)<0) and x1 <= imshape[1]/2 and x2 <= imshape[1]/2: 
                         left_line = np.append(left_line,[[x1,x2],[y1,y2]],axis=1) 
                elif ((y1-y2)/(x1-x2+1e-4)>0) and x1 >= imshape[1]/2 and x2 >= imshape[1]/2: 
                         right_line = np.append(right_line,[[x1,x2],[y1,y2]],axis=1) 
    
    if len(left_line[1]) > 0 and s_flag_l == 0:
        left_fit = np.polyfit(left_line[1], left_line[0],1)
        if left_fit[0] <0 and abs(left_fit[0])<(imshape[1]/imshape[0]*4):
        #if left_fit[0] <0:
            s_flag_l = 1
            line_image = cv2.line(img, (int(np.polyval(left_fit, imshape[0])),imshape[0]),(int(np.polyval(left_fit, imshape[0]*0.6)),int(imshape[0]*0.6)), color, thickness)
    if len(right_line[1]) > 0 and s_flag_r == 0:
        right_fit = np.polyfit(right_line[1], right_line[0],1)
        #if right_fit[0] >0:
        if right_fit[0] >0 and abs(right_fit[0])<(imshape[1]/imshape[0]*4):
            s_flag_r = 1
            line_image = cv2.line(img, (int(np.polyval(right_fit, imshape[0])),imshape[0]),(int(np.polyval(right_fit, imshape[0]*0.6)),int(imshape[0]*0.6)), color, thickness)
    
    return s_flag_l,s_flag_r
   
    
def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap, s_flag_l, s_flag_r):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]) , minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    [s_flag_l, s_flag_r] = draw_lines(line_img, lines, s_flag_l, s_flag_r)
    return line_img, s_flag_l, s_flag_r


# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=1, β=1., γ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + γ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, γ)
```

## Pipeline

**Step 1:** Covert all the images from RGB to HSL. 

```python
def to_hsl(img):
    return cv2.cvtColor(img, cv2.COLOR_RGB2HLS)


hsl_images = list(map(lambda img: to_hsl(img), original_images))
vis_image(hsl_images,original_image_names)
```

**Step 2:** Filter only white and yellow color (the two colors of the lane)

```python
def filter_yellow_and_white_hsl(img):
    #filter out non-yellow color
    low_threshold_yellow = np.array([15, 38, 115], dtype=np.uint8)
    high_threshold_yellow = np.array([35, 204, 255], dtype=np.uint8)  
    yellow_mask = cv2.inRange(img, low_threshold_yellow, high_threshold_yellow)
    
    #filter out non-white color
    low_threshold_white = np.array([0, 200, 0], dtype=np.uint8)
    high_threshold_white = np.array([180, 255, 255], dtype=np.uint8) 
    white_mask = cv2.inRange(img, low_threshold_white, high_threshold_white)
    
    #combine the two filtered results
    final_filter = cv2.bitwise_or(yellow_mask,white_mask)
    img = cv2.bitwise_and(img,img,mask=final_filter)
    return img


filter_images = list(map(lambda img: filter_yellow_and_white_hsl(img), hsl_images))
vis_image(filter_images,original_image_names)
```

**Step 3**: Perform gray scale conversion, canny fitler, and regional filter as suggested in the lecture

```python
def gary_canny_regional_fitler(img,high_threshold=200):
        #Convert the image to gray scale and apply Gaussian smoothing
        gray_image = grayscale(img)
        kernel_size = 5
        gray_image = gaussian_blur(gray_image,kernel_size)
        
        
        #Apply Canny filter on the image
        low_threshold = high_threshold/2
        canny_image=canny(gray_image, low_threshold, high_threshold)

        #Apply regional filter to the image
        imshape = canny_image.shape
        vertices = np.array([[(0,imshape[0]),(imshape[1]*0.5,imshape[0]*0.6), (imshape[1]*0.5,imshape[0]*0.6), (imshape[1],imshape[0])]],dtype = np.int32)
        final_image = region_of_interest(canny_image, vertices)
        
        return final_image

print('First, we use try high threshold = 400 in the canny function')
gary_canny_regional_images = list(map(lambda img: gary_canny_regional_fitler(img,high_threshold=400), filter_images))
vis_image(gary_canny_regional_images,original_image_names)
plt.show()

print('Second, we use try high threshold = 200 in the canny function')
gary_canny_regional_images = list(map(lambda img: gary_canny_regional_fitler(img,high_threshold=200), filter_images))
vis_image(gary_canny_regional_images,original_image_names)
plt.show()
```

We did not see a substantial difference between threshold = 400 and 200. So we will stick with 200 for now.

**Step 4**: Apply Hough transormation to find lines (a while loop to optmize the function)

```python
def line_finding(img):
    max_line_gap = 20
    min_line_len = 1
    threshold = 10
    rho = 1 # distance resolution in pixels of the Hough grid
    theta = np.pi/180 # angular resolution in radians of the Hough grid
    s_flag_l=0
    s_flag_r=0
    while s_flag_l * s_flag_r == 0:
        [line_image,s_flag_l,s_flag_r] = hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap, s_flag_l, s_flag_r)
        if s_flag_l == 1 or s_flag_r ==1:
            final_image = weighted_img(line_image, line_image) 
        min_line_len = min_line_len-2
        threshold = threshold-1
        max_line_gap = max_line_gap + 1
    return final_image

line_images = list(map(lambda img: line_finding(img), gary_canny_regional_images))
vis_image(line_images,original_image_names)
```

**Step 5**: piplline all the functions

```python
def process_image(img):
    line_image = line_finding(gary_canny_regional_fitler(filter_yellow_and_white_hsl(to_hsl(img))))
    img = weighted_img(line_image, img)     
    return img
```
